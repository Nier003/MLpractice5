### Visualization with Seaborn

<div style="direction:rtl">
ایمپورت کتابخانه ها.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

<div style="direction:rtl">
ست کردن استایل.

```python
sns.set_style('whitegrid')
```

<div style="direction:rtl">   
لود کردن دیتا.

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.jointplot(y=titanic['age'],x=titanic['fare'])
```

<div style="direction:rtl">
نمودار کرایه بر حسب سن را به شکل زیر رسم نمایید:
<div style="direction:rtl">
راهنمایی: نام ستون ترسیمی در کنار محور ها درج شده است.

<div style="direction:rtl">
نمودار توزیع آماری کرایه ها (distplot) را به شکل رسم نمایید:
<div style="direction:rtl">
پ.ن: وارنینگ نمایش داده شده اهمیتی ندارد و مشکلی وجود ندارد.

```python
sns.distplot( titanic['fare'] , bins=30 , kde=False , color='red')
```

<div style="direction:rtl">
نمودار جعبه ای کلاس بلیت را بر حسب سن رسم نمایید.

```python
sns.boxplot(x='class', y='age' , data= titanic , palette='rainbow')
```

<div style="direction:rtl">
همان نمودار کلاس بر حسب سن  را با نمود تجمع افراد در سن های مختلف رسم نمایید.
<div style="direction:rtl">
پ.ن: وارنینگ نمایش داده شده اهمیتی ندارد و مشکلی وجود ندارد.

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
```

<div style="direction:rtl">
نمودار تعداد افراد حاضر در کشتی را بر حسب جنسیت رسم نمایید.

```python
sns.countplot(x = 'sex'  , data = titanic)
```

<div style="direction:rtl">
نمودار میزان همبستگی را در دیتاست تایتانیک رسم کنید و برای آن عنوان Titanic Correlations قرار دهید.
<div style="direction:rtl">
پ.ن: وارنینگ نمایش داده شده اهمیتی ندارد و مشکلی وجود ندارد.

```python
sns.heatmap(titanic.corr(),cmap='Greens')
```

<div style="direction:rtl">
نمودار توزیع آماری سن بر حسب جنسیت را در دو شبکه ی جدا با کمک سیبورن رسم نمایید.

<div style="direction:rtl">   
نمودار نقطه ای دو ستون a و b را ترسیم نماید به شکلی که سایز نقاط آن50، رنگ آن قرمز و سایز قاب آن 12 در 3 باشد.

```python
df3.plot(x='a', y='b',color = 'r',s = 50, kind="scatter", figsize=(12,3))
```

<div style="direction:rtl">   
نمودار توزیع آماری ستون a را رسم نمایید.

```python
df3["a"].plot(kind = 'hist')
```

<div style="direction:rtl">   
نمودار توزیع آماری ستون a را با استایل ggplot و تعداد ستون 25 و شفافیت(alpha) 0.5 رسم نمایید.

```python
plt.style.use('ggplot')
```

```python
#با توجه به تصویر 2تا پایین(نمودار اصلی) تعداد ستون را 50 انتخاب کردم 
df3["a"].plot(kind = 'hist',bins = 50 , color = 'r', alpha = 0.5)
```

<div style="direction:rtl">   
نمودار جعبه ای ستون های a و b را رسم نمایید.
<div style="direction:rtl">   
نکته ی مهم : دقت فرمایید که در پانداز، نموداری جعبه ای را می توان برای خود ستون ترسیم نمود در حالی که در سیبورن یک ستون کتگوریکال لازم بود.

```python
df3[['a','b']].boxplot()
```

<div style="direction:rtl">   
تابع چگالی احتمال را برای ستون d ترسیم کنید.

```python
df3['d'].plot.kde()
```

<div style="direction:rtl">   
همان تابع بالا را به شکل مقطع و با عرض 5 رسم نمایید.

```python
df3['d'].plot.kde(lw = 5,ls = '--')
```
